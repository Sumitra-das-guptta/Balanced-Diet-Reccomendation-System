```python
#### import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
import sklearn
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import scale
import sklearn.metrics as sm
from sklearn import datasets
from sklearn.metrics import confusion_matrix,classification_report

from sklearn.neighbors import KNeighborsClassifier
```

```python
dataset = pd.read_csv('balanced diet.csv')
```

```python
X= dataset.iloc[:,[0,2]].values
```

```python
X[1]
```

```python
wcss=[]
```

```python
for i in range(1,15):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plot.plot(range(1,15),wcss)
plot.title('Elbow')
plot.xlabel('no of clusters')
plot.ylabel('wcss')
```

```python
 kmeans=KMeans(n_clusters=4,init='k-means++',random_state=0)
y = kmeans.fit_predict(X)
```

```python

y
```

```python
plot.scatter(X[y == 0, 0], X[y == 0,1],s = 100, c='red', label = 'Cluster 1')

plot.scatter(X[y == 1, 0], X[y == 1,1],s = 100, c='blue', label = 'Cluster 2')

plot.scatter(X[y == 2, 0], X[y == 2,1],s = 100, c='green', label = 'Cluster 3')

plot.scatter(X[y == 3, 0], X[y == 3,1],s = 100, c='cyan', label = 'Cluster 4')

plot.title('kmeans clustering')
plot.xlabel('age')
plot.ylabel('bmi')
plot.legend()
plot.show()
```
